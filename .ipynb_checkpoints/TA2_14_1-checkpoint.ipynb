```python
import pandas as pd
from IPython.display import display
import numpy as np

from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
data_kelompok14=pd.read_csv('./Dataset_14_1.csv')
pd.set_option('display.max_row', None)
display(data_kelompok14)
```

```python
data14 = data_kelompok14.drop(columns=['Target'])
```

```python
x=data14.values 
x=np.delete(x,4,axis=1) 
y=data14['Skor Pelanggaran'].values
```

```python
x
```

```python
y
```

```python
from sklearn.naive_bayes import GaussianNB
modelnb = GaussianNB()
```

```python
def get_score(modelnb,X_train, X_test, y_train, y_test):
    modelnb.fit(X_train,y_train)
    return modelnb.score(X_test,y_test)
```

```python
kf=KFold(n_splits=5, random_state=123, shuffle=True)
print(kf)
scores_nb = []
i=1
for train_index, test_index in kf.split(x,y):
    print("Fold ", i)
    print("TRAIN :", train_index, "TEST :", test_index)
    X_train=x[train_index]
    X_test=x[test_index]
    y_train=y[train_index]
    y_test=y[test_index]
    i+=1
    scores_nb.append(get_score(GaussianNB(), X_train, X_test, y_train, y_test))
    print ("Train set:",X_train.shape, y_train.shape)
    print ("Test set:",X_test.shape, y_test.shape)
```

```python
print("akurasi per fold :")
scores_nb
```

```python
prediksi = modelnb.predict(X_test)
prediksi
```

```python
# import confusion_matrix model
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, prediksi)
```

```python
# Menghitung nilai akurasi dari klasifikasi naive bayes 
from sklearn.metrics import classification_report
print(classification_report(y_test, prediksi))
```
